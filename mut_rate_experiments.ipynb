```python
from custom_pool import CustomPool
```

```python
pool_size = 5
runs = 25
mut_rates = (None, 0.2, 0.4, 0.6, 0.8)
print(mut_rates)
```

```python
def experiment_fn(mut_rate):
    from run import experiment
    
    params = {
        "mut_rate": mut_rate,
    }
    
    exp_data = experiment(params, runs)
    
    return mut_rate, exp_data
```

```python
exp_results = None

with CustomPool(pool_size) as pool:
    res = pool.map_async(experiment_fn, mut_rates)
    
    exp_results = res.get()
```
