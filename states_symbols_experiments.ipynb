```python
import itertools

from custom_pool import CustomPool
```

```python
pool_size = 4
runs = 20
sizes = [3, 5, 10, 25]
param_pairs = list(itertools.product(sizes, sizes))
param_pairs += [(50, 3), (50, 5), (50, 10), (50, 25)]
param_pairs += [(100, 3), (100, 5), (100, 10), (100, 25)]
print(param_pairs)
```

```python
def experiment_fn(param_pair):
    from run import experiment
    
    (states, symbols) = param_pair
    params = {
        "states": states,
        "symbols": symbols,
    }
    
    exp_data = experiment(params, runs)
    
    return param_pair, exp_data
```

```python
exp_results = None

with CustomPool(pool_size) as pool:
    res = pool.map_async(experiment_fn, param_pairs)
    
    exp_results = res.get()
```
